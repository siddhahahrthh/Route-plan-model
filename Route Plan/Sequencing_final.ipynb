```python
import pandas as pd
import folium
import requests
from geopy.distance import geodesic
import gspread


def distance(origin , destination):
    url = f"https://dev.virtualearth.net/REST/v1/Routes/DistanceMatrix?key=AriqFuvNl9dRgbk3qtJd9635LWW_O4n4Kx0OZj6lWNdnTTVynJCGXoqUi0jpQ8zT&origins={origin}&destinations={destination}&travelMode=driving"
    response = requests.get(url)
    data = response.json()
    return data['resourceSets'][0]['resources'][0]['results'][0]['travelDistance']
def time(origin , destination):
    url = f"https://dev.virtualearth.net/REST/v1/Routes/DistanceMatrix?key=AriqFuvNl9dRgbk3qtJd9635LWW_O4n4Kx0OZj6lWNdnTTVynJCGXoqUi0jpQ8zT&origins={origin}&destinations={destination}&travelMode=driving"
    response = requests.get(url)
    data = response.json()
    return data['resourceSets'][0]['resources'][0]['results'][0]['travelDuration']
def routecalculation(route , coordinates , time_taken , start , data , userids , coordinate , orderids , region , rider , number , slot , address):
    while coordinates:
        nearest = min(coordinates, key=lambda x: distance(start, x))
        user = coordinate.loc[coordinate['Coordinates'] == nearest , 'lng'].iloc[0]
        usd = data.loc[data['lng'] == user , 'user_id'].iloc[0]
        userids.append(usd)
        orderids.append(data.loc[data['lng'] == user , 'order_id'].iloc[0])
        rider.append(data.loc[data['lng'] == user , 'Rider Assigned Final'].iloc[0])
        region.append(data.loc[data['lng'] == user , 'Cluster Name'].iloc[0])
        address.append(data.loc[data['lng'] == user , 'address'].iloc[0])
        number.append(data.loc[data['lng'] == user , 'mobile'].iloc[0])
        slot.append(data.loc[data['lng'] == user , 'updated_delivery_time'].iloc[0])
        data = data[data['user_id'] != usd]
        time_taken.append(str(time(start , nearest)) + " minutes")
        coordinates.remove(nearest)
        start = nearest
    return start




sheet_url = 'https://docs.google.com/spreadsheets/d/1TXJYYulF8ZHjC_1iV7_Efjb6GyC3JFjis8vKQU80Jwc/edit#gid=962846952'
sheet_name = '2.Raw - Sorted'
dest_name = 'Delivery Sequence'
gc = gspread.service_account(filename='big.json')
sh = gc.open_by_url(sheet_url)
worksheet = sh.worksheet(sheet_name)
destsheet = sh.worksheet(dest_name)
last = len(destsheet.get_all_values()) +1 
cell_range = f'A{last}'
sh = worksheet.get_all_values()
df = pd.DataFrame(sh[1:], columns=sh[0])
drivers = df['Rider Assigned Final'].unique()
for driver in drivers:
    data = df[df['Rider Assigned Final'] == driver]
    warehouse = "28.43329932748821,77.0800348638816"
    start = warehouse
    userids = []
    orderids = []
    route = [warehouse]
    time_taken = []
    region = []
    slot = []
    address = []
    rider = []
    number = []
    for i in range(6,12):
        coordinates = data.loc[data['updated_delivery_time'] == f'Morning {i}am - {i+1}am', ['lat', 'lng']]
        coordinates['Coordinates'] = coordinates['lat'].astype(str) + ',' + coordinates['lng'].astype(str)
        coordinates_list = coordinates['Coordinates'].tolist()
        if(len(coordinates_list)>0):
            start = routecalculation(route , coordinates_list , time_taken , start , data , userids , coordinates , orderids , region , rider , number , slot , address)
    
    coordinates = [tuple(map(float, coord.split(','))) for coord in route]
    route.remove(warehouse)
    new_df = pd.DataFrame({"user id": userids , "order id": orderids , "coordinates": rider, "slot" : slot , "number" :number , "address" : address , "region" : region ,"time_duration": time_taken})
    cell_range = f'A{last}'
    destsheet.update(values=new_df.values.tolist() , range_name = cell_range)
    last = last + len(userids)
        
print("done") 
```
